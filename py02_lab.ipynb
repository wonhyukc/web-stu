## Section 1: print() 완전 정복

### 1-1. print() 기본 출력

`print()` 함수는 괄호 안의 내용을 화면에 출력합니다.  
아래 코드를 실행한 다음, 직접 원하는 문장으로 바꿔서 실행해 보세요.

```python
# 문자열 출력
print("안녕하세요, 파이썬!")

# 숫자 출력 (따옴표 없이)
print(42)
print(3.14)

# 여러 값 동시 출력 (콤마로 구분)
print("이름:", "홍길동", "나이:", 20)
```

### 1-2. 콤마(`,`) vs 더하기(`+`) 비교

두 방식의 핵심 차이를 직접 눈으로 확인하세요.

| 방법 | 공백 | 타입 혼합 |
|:-----|:----:|:--------:|
| 콤마(`,`) | 자동 삽입 | 가능 |
| `+` 연결 | 직접 삽입 필요 | 문자열끼리만 가능 |

```python
# 콤마: 자동 공백, 타입 혼합 가능
print("이름:", "홍길동")
print("나이:", 20)
print("이름:", "홍길동", "나이:", 20)
```

```python
# + 연결: 공백 없이 바짝 붙음, 문자열끼리만
print("안녕" + "하세요")
print("안녕 " + "하세요")  # 공백은 직접 넣어야 함

# 문자열 반복도 가능
print("-" * 30)
print("하하" * 3)
```

```python
# TypeError 발생 예시 — 문자열 + 숫자는 + 로 연결 불가
# 에러 메시지를 읽고 어떤 내용인지 파악해 보세요.
나이 = 20
print("나의 나이는 " + 나이 + "살입니다.")  # 에러 발생!
```

### 1-3. end= 옵션과 sep= 옵션

- `end=` : 출력 끝에 붙는 문자를 바꿉니다 (기본값: 줄바꿈 `"\n"`)
- `sep=` : 콤마로 나열한 값들 사이의 구분자를 바꿉니다 (기본값: 공백 `" "`)

```python
# end= 옵션: 출력 끝을 바꾼다
print("로딩 중", end="...")
print("완료!")

# 줄바꿈 없이 이어서 출력
print("A", end="")
print("B", end="")
print("C")

# 월~금 한 줄 출력
print("월", end=" ")
print("화", end=" ")
print("수")
```

```python
# sep= 옵션: 값 사이 구분자를 바꾼다
print("2026", "03", "07", sep="-")          # 날짜 형식
print("사과", "바나나", "딸기", sep=", ")    # 목록 형식
print("A", "B", "C", sep=" → ")            # 흐름 표시

# end= 와 sep= 동시 사용
print("사과", "바나나", sep=" | ", end="!\n")
```

`#` 뒤에 오는 텍스트는 파이썬이 실행할 때 완전히 무시합니다.  
코드 설명을 달거나, 특정 줄을 임시로 비활성화할 때 사용합니다.

## Section 2: 4가지 기본 타입 + 산술 연산자

### 2-1. 4가지 기본 타입

파이썬의 기본 자료형 4가지:

| 타입 | 이름 | 예시 | 특징 |
|:----:|:-----|:-----|:-----|
| `str` | 문자열 | `"안녕"`, `"42"` | 따옴표로 감싼 것 |
| `int` | 정수 | `42`, `-5`, `0` | 소수점 없는 숫자 |
| `float` | 부동소수점 | `3.14`, `175.5` | 소수점 있는 숫자 |
| `bool` | 불린 | `True`, `False` | 참/거짓 두 값만 |

```python
# str — 따옴표 안에 있으면 문자열
이름 = "홍길동"
전화번호 = "010-1234-5678"  # 숫자처럼 보이지만 문자열!
print(이름)
print(전화번호)

# "42"와 42는 전혀 다릅니다!
print("42" + "8")   # 428  (문자열 연결)
print(42 + 8)        # 50   (숫자 덧셈)
```

```python
# int — 소수점 없는 숫자
나이 = 20
온도 = -3
print("나이:", 나이, "온도:", 온도)

# float — 소수점 있는 숫자
키 = 175.5
원주율 = 3.14159
print("키:", 키, "원주율:", 원주율)

# 주의: 일반 나눗셈 / 의 결과는 float!
print(10 / 2)    # 5.0  (int가 아님)
```

```python
# bool — True 또는 False 두 값만 존재
# 반드시 첫 글자가 대문자여야 합니다.
로그인_상태 = True
시험_합격 = False
print(로그인_상태, 시험_합격)

# 비교 연산의 결과도 bool
print(10 > 5)     # True
print(10 < 5)     # False
print(10 == 10)   # True
print(3 == 5)     # False
```

### 2-2. type() 함수로 타입 확인하기

`type(값)` 을 사용하면 어떤 데이터의 타입인지 즉시 확인할 수 있습니다.  
헷갈릴 때마다 `print(type(값))`으로 확인하는 습관을 들이세요.

```python
# 기본 타입 확인
print(type("안녕하세요"))  # <class 'str'>
print(type(42))            # <class 'int'>
print(type(3.14))          # <class 'float'>
print(type(True))          # <class 'bool'>
```

```python
# 헷갈리는 경우 직접 확인
print(type("42"))   # str  — 따옴표 안이므로 문자열
print(type(42))     # int  — 따옴표 없는 정수
print(type(42.0))   # float — 소수점이 있으므로 부동소수점

# 나눗셈 결과 타입 확인
결과 = 10 / 2
print(결과)           # 5.0
print(type(결과))     # <class 'float'>

# 직접 확인해 보기
x = "100"
y = 100
z = True
print(type(x))   # 주석을 해제하지 않아도 됩니다, 이미 실행됩니다
print(type(y))
print(type(z))
```

### 2-3. 산술 연산자 7종

| 연산자 | 이름 | 예시 (`a=17, b=5`) | 결과 |
|:------:|:-----|:-------------------|:----:|
| `+` | 더하기 | `a + b` | `22` |
| `-` | 빼기 | `a - b` | `12` |
| `*` | 곱하기 | `a * b` | `85` |
| `/` | 나누기 | `a / b` | `3.4` |
| `//` | 정수 나누기 (몫) | `a // b` | `3` |
| `%` | 나머지 | `a % b` | `2` |
| `**` | 거듭제곱 | `2 ** 10` | `1024` |

```python
a = 17
b = 5

print("더하기:", a + b)
print("빼기:", a - b)
print("곱하기:", a * b)
print("나누기:", a / b)           # 결과는 float
print("정수 나누기(몫):", a // b)
print("나머지:", a % b)
print("거듭제곱(2의 10제곱):", 2 ** 10)
```

```python
# // 와 % 활용: 빵 나눠먹기
빵_개수 = 17
사람_수 = 5
print("한 사람당 받는 빵:", 빵_개수 // 사람_수, "개")
print("남은 빵:", 빵_개수 % 사람_수, "개")

# % 로 짝수/홀수 판별
숫자 = 15
print(숫자, "% 2 =", 숫자 % 2)   # 1 이면 홀수

# ** 로 거듭제곱
print("2의 3제곱:", 2 ** 3)    # 8
print("3의 2제곱:", 3 ** 2)    # 9

# 연산자 우선순위: 괄호로 명확하게 표현하는 습관
print(2 + 3 * 4)      # 14 (곱하기 먼저)
print((2 + 3) * 4)    # 20 (괄호 안 먼저)
```

## Section 3: input() + 타입 변환 + 에러

### 3-1. input() 함수: 사용자 입력 받기

`input()` 함수는 실행 중에 잠깐 멈추고 사용자 입력을 기다립니다.  
Enter를 누르면 입력한 값을 반환합니다.

> **핵심**: `input()`은 **항상 `str`(문자열)을 반환**합니다.  
> 숫자를 입력해도 문자열로 받아집니다.

```python
# 이름 입력받아 출력
이름 = input("이름을 입력하세요: ")
print("안녕하세요,", 이름, "님!")
```

```python
# 숫자를 입력해도 str로 반환됨 — 직접 확인!
값 = input("숫자를 입력하세요: ")
print("입력한 값:", 값)
print("타입:", type(값))   # <class 'str'> 가 나옵니다
```

### 3-2. 타입 변환: int(), float(), str()

| 함수 | 방향 | 예시 | 결과 |
|:----:|:-----|:-----|:----:|
| `int()` | str/float → 정수 | `int("42")` | `42` |
| `float()` | str/int → 소수 | `float("3.14")` | `3.14` |
| `str()` | int/float → 문자열 | `str(100)` | `"100"` |

```python
# int() 변환 후 계산
나이 = int(input("나이를 입력하세요: "))
다음_나이 = 나이 + 1
print("내년에는", 다음_나이, "살입니다.")
print("변환 후 타입:", type(나이))   # <class 'int'>
```

```python
# float() 변환
키 = float(input("키를 입력하세요 (cm): "))
print("키는", 키, "cm입니다.")
print("변환 후 타입:", type(키))   # <class 'float'>
```

```python
# str() 변환 — 숫자를 + 로 문자열에 연결할 때 필요
나이 = 20
문장 = "나의 나이는 " + str(나이) + "살입니다."
print(문장)

# int(float()) 이중 변환
print(int(float("3.14")))   # 3
```

### 3-3. SyntaxError vs TypeError — 에러를 직접 만들어 보자

| 에러 | 발생 시점 | 원인 |
|:-----|:--------:|:-----|
| `SyntaxError` | 실행 전 | 파이썬 문법 오류 (괄호 누락, 오타 등) |
| `TypeError` | 실행 중 | 타입 충돌 (str + int 등) |

아래 셀들을 실행해서 두 에러 메시지의 차이를 직접 눈으로 확인하세요.  
에러 메시지의 **맨 마지막 줄**이 핵심 내용입니다.

```python
# SyntaxError 유발 — 따옴표를 닫지 않은 경우
# 에러 메시지를 읽고 무엇이 문제인지 파악해 보세요.
# 그다음 AI에게 이 에러 메시지를 복사해서 질문해 보세요.
print("안녕하세요)
```

```python
# TypeError 유발 — str + int 연산 시도
# input()의 반환값은 str이므로, 변환 없이 숫자를 더하면 에러가 납니다.
나이 = input("나이를 입력하세요: ")   # 반환값은 str
다음_나이 = 나이 + 1                   # str + int → TypeError!
```

```python
# TypeError 수정 — int()로 변환해서 해결
나이 = int(input("나이를 입력하세요: "))  # int()로 감싸면 해결
다음_나이 = 나이 + 1
print("내년 나이:", 다음_나이)
```
